The following functions read the original datasets into colab. Since there is one Tweet dataset file for each day, I have to read all the files separately and concat them in the end. Reading all the Coronavirus-related Tweets in March







```
twitter_march = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Dataset/2020-03-00 Coronavirus Tweets (pre 2020-03-12).CSV')
twitter_march = twitter_march.loc[twitter_march['is_quote']==False].drop_duplicates()
twitter_march = twitter_march.drop(['status_id', 'screen_name', 'reply_to_status_id', 'reply_to_user_id', 'reply_to_screen_name', 'account_lang', 'is_quote', 'is_retweet'], axis=1)

for i in range(12,32):
    dataset = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Dataset/2020-03-'+str(i)+' Coronavirus Tweets.CSV')
    dataset = dataset.loc[(dataset['is_quote']==False)].drop_duplicates()
    dataset = dataset.drop(['status_id', 'screen_name', 'reply_to_status_id', 'reply_to_user_id', 'reply_to_screen_name', 'account_lang', 'is_quote', 'is_retweet'], axis=1)
    twitter_march = pd.concat([twitter_march, dataset])
```

Reading all the Coronavirus-related Tweets in April

```
twitter_april = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Dataset/2020-04-01 Coronavirus Tweets.CSV')
twitter_april = twitter_april.loc[twitter_april['is_quote']==False].drop_duplicates()
twitter_april = twitter_april.drop(['status_id', 'screen_name', 'reply_to_status_id', 'reply_to_user_id', 'reply_to_screen_name', 'account_lang', 'is_quote', 'is_retweet'], axis=1)

for i in range(2,31):
  if i < 10:
    i = '0'+str(i)
    dataset = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Dataset/2020-04-'+i+' Coronavirus Tweets.CSV')
    dataset = dataset.loc[(dataset['is_quote']==False)].drop_duplicates()
    dataset = dataset.drop(['status_id', 'screen_name', 'reply_to_status_id', 'reply_to_user_id', 'reply_to_screen_name', 'account_lang', 'is_quote', 'is_retweet'], axis=1)
    twitter_april = pd.concat([twitter_april, dataset])
  else:
    dataset = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Dataset/2020-04-'+str(i)+' Coronavirus Tweets.CSV')
    dataset = dataset.loc[(dataset['is_quote']==False)].drop_duplicates()
    dataset = dataset.drop(['status_id', 'screen_name', 'reply_to_status_id', 'reply_to_user_id', 'reply_to_screen_name', 'account_lang', 'is_quote', 'is_retweet'], axis=1)
    twitter_april = pd.concat([twitter_april, dataset])
```

Final full dataset: all the Coronavirus-related Tweets in March and April

```
twitter = pd.concat([twitter_march, twitter_april])
```

### Preparing for country & time series analysis dataset

The country & time series analysis part of the project explores the relationships between countries, Tweets, and Coronavirus cases. However, given that more than 95% of the data entries have missing values for "country" column, I decided to drop them and only keep the data entries with values for "country" in `twitter_country` dataset for the analysis of this part, which has around 1 million entries.

```
twitter_country = twitter.dropna(subset=['country_code']).reset_index(drop=True)
twitter_country
```

Since the dataset only has "country_code" rather than the names of the countries, which could be difficult to understand, I created a new column "country" by using the `countries_dictionary` provided by the author to explain what each country_code means for future analysis.

```
countries_dictionary = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/Countries.CSV')
```

```
countries_dictionary = countries_dictionary.set_index(['country_code']).to_dict()
countries_dictionary = countries_dictionary['country']
countries_dictionary.update({'FM': 'Federated States of Micronesia'})
countries_dictionary.update({'NR': 'Nauru'})
```

```
twitter_country["country"] = twitter_country["country_code"].map(lambda x: countries_dictionary[x])
```

```
twitter_country
```

Export the `twitter_country` dataset for future usage.

```
twitter_country.to_csv('twitter_country.csv', quoting=csv.QUOTE_NONNUMERIC, index = False)
```

To add information regarding Coronavirus cases into the analysis, I obtained the `time_series_covid19_confirmed_global.csv` dataset from John Hopkins, which has the data for confirmed Coronavirus cases every day for each country.

```
coronavirus = pd.read_csv('/content/gdrive/My Drive/PENN/2020 Spring/CIS 545 Big Data Analytics/Final Project/time_series_covid19_confirmed_global.csv')
```

Next, I dropped the columns that are not relevant, including latitutde, longitude, and dates in May, which don't have any Twitter data.

```
coronavirus = coronavirus.drop(['Province/State', 'Lat', 'Long', '5/1/20', '5/2/20'], axis=1)
coronavirus = coronavirus[['Country/Region']].join(coronavirus.iloc[:, 43: ])
```

Then, I created a new dataset `coronavirus_time` that have the total number of confirmed Coronavirus cases every day globally. This will be used for the time series analysis to observe the global trend of Tweet counts and number of confirmed coronavirus cases. 

I also changed the date to the right format 'month-day" (e.g. 03-01) for future analysis. 

```
coronavirus_time = pd.DataFrame(coronavirus.sum(axis=0), index=None).reset_index()[1:]
coronavirus_time.columns = ['Date', 'Number of Cases']
coronavirus_time["Date"] = pd.to_datetime(coronavirus_time["Date"])
coronavirus_time["Date"] = coronavirus_time["Date"].dt.strftime('%y-%m-%d').apply(lambda x: x[3:])
```

```
coronavirus_time
```

Export the `coronavirus_time` dataset for future usage.

```
coronavirus_time.to_csv('coronavirus_time.csv', quoting=csv.QUOTE_NONNUMERIC, index = False)
```

Next, I created a new dataset `coronavirus_country` that have the total number of confirmed Coronavirus cases for every country until 04-30. This will be used for the country analysis to observe the relationship between Tweet counts and the number of Coronavirus cases by country. 

```
coronavirus_country = pd.DataFrame(coronavirus.sum(axis=1))
coronavirus_country.columns = ['Number of Cases']
coronavirus_country = coronavirus[['Country/Region']].join(coronavirus_country)
```

I also modified the "Country/Region" values in the dataset by creating a `name_change_dictionary` so that the names of the countries/regions will match the names of the countries/regions in the Twitter dataset for the future analysis.

```
name_change_dictionary = {'Philippines': 'Republic of the Philippines', 'Iran': 'Islamic Republic of Iran', 'Netherlands': 'The Netherlands',  
                          'Saudi Arabia': 'Kingdom of Saudi Arabia', "China": "People's Republic of China", "Slovenia": 'Republic of Slovenia', 
                          'Taiwan*':'Taiwan', 'Czechia': 'Czech Republic', 'Jordan':'Hashemite Kingdom of Jordan', 'Croatia':'Republic of Croatia', 
                          'Congo (Brazzaville)': 'Congo Brazzaville', 'Serbia': 'Republic of Serbia', 'Mauritius': 'Republic of Mauritius', 
                          'Congo (Kinshasa)': 'Democratic Republic of Congo', 'Timor-Leste':'East Timor', 'Slovakia':'Slovak Republic', 
                          'North Macedonia':'Former Yugoslav Republic of Macedonia', 'US': 'United States', 'Korea, South':'Republic of Korea'}
```

```
coronavirus_country['Country/Region'] = coronavirus_country['Country/Region'].apply(lambda x: name_change_dictionary[x] if x in name_change_dictionary.keys() else x)
```

```
coronavirus_country
```

Export the `coronavirus_country` dataset for future usage.

```
coronavirus_country.to_csv('coronavirus_country.csv', quoting=csv.QUOTE_NONNUMERIC, index = False)
```

### Preparing for modeling dataset

The purpose of the model in this project is to predict the influence of Tweets, i.e., the number of retweets of Tweets. While the original Twitter dataset has Tweets in 65 different languages, this model will only predict the influence of Tweets in English language for practical reasons. This is because while the features of the model will include certain textual features of Tweets, such as the length of texts, it could be very difficult and unreasonable to do textual analysis for different languages and use them as features in the model. 

I selected all Tweets in English from the original dataset as the modeling dataset `twitter_en`, which has around 13 million entries.

```
twitter_en = twitter.loc[twitter['lang']=='en']
twitter_en
```

Export the exploratory dataset `twitter_en` for future usage.

```
twitter_en.to_csv('twitter_en.csv', quoting=csv.QUOTE_NONNUMERIC, index = False)
```
